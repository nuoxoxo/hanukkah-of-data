```python
import pandas as pd
A = pd.read_csv('noahs-customers.csv')
def abbrev (name):
    res = ''
    for pt in name.split():
        res += pt[0]
    return res
A.name = A.name.apply(abbrev)
A[A.name == 'JP'].head()
```
