```python
import pandas as pd
import os
Headers = [_.split('.')[0] for _ in os.listdir('./') if '.csv' in _]
Headers
# Got all the CSV filenames w/o extensions
```

```python
CSVS = [pd.read_csv(header + '.csv') for header in Headers]
len( CSVS )
# Should be 4 as there are 4 files in total
```

```python
NO, NP, NOIT, NC = CSVS
[len(_) for _ in [NO, NP, NOIT, NC]]
# Final setup: we need all 4 DataFrames
```

```python
def abbrev (name):
    return ''.join(_[0] for _ in name.split()).upper() == 'JP'

JP = NC[NC.name.apply(lambda x: abbrev(x))]
JP
# Got all contractors whose initials == JP
```

```python
orders17 = NO[NO['ordered'].str.startswith('2017')]#\
    # .sort_values('ordered', ascending=False)\
    # .groupby('customerid')\
    # .first()\
    # .ordered
# (alt. to find 2017 : NO[pd.to_datetime(NO['ordered']).dt.year == 2017])

orders17 # most_recent_order
# List of orders made back in 2017 reverse-sorted
```

```python
JP17 = pd.merge(orders17, JP, on='customerid')
JP17.name.value_counts()
```

```python
orders17last = orders17.sort_values('ordered', ascending=False)\
    .groupby('customerid')\
    .first()\
    .ordered

orders17last
```

```python
JP17last = pd.merge(orders17last, JP17, on="customerid")
JP17last
```

```python
Coffee = NP[NP['desc'].str.contains('coffee', case=False)]
Coffee
```

```python
Coffee_order = pd.merge(Coffee, NOIT, on='sku')
Coffee_order
```

```python
Coffee_order_details = pd.merge(Coffee_order, JP17last, on='orderid')
Coffee_order_details
# Coffee_order.head(), JP17.head()
```
