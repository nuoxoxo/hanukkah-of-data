```python
import pandas as pd
import os
filenames = [_.split('.')[0] for _ in os.listdir('./') if '.csv' in _]
print(filenames)

dfs = [pd.read_csv(filename + '.csv') for filename in filenames]
df_orders, df_prod, df_order_items, df_customers = dfs
print([len(_) for _ in [df_orders, df_prod, df_order_items, df_customers]])
```

```python
cat_prod = df_prod[ df_prod['desc'].str.contains('cat',case=False) ]
cat_prod
```

```python
cat_items = pd.merge(cat_prod, df_order_items, on='sku')
cat_items
```

```python
cat_orders = pd.merge(cat_items, df_orders, on='orderid')
cat_orders
```

```python
cat_clients = pd.merge(cat_orders, df_customers, on='customerid')
cat_clients
```

```python
cat_clients[ cat_clients.citystatezip.str.contains('Staten', case=False) ]
    .groupby(['name','phone'])\
    .size()\
    .sort_values(ascending=False)\
    .head(4)
```
