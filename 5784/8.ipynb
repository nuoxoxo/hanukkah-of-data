```python
import pandas as pd
customers_df = pd.read_csv('noahs-customers.csv')
# sample on customerid==6878 
customers_df[ customers_df['customerid'] == 6878 ]
```

```python
orders_df = pd.read_csv('noahs-orders.csv')
orders_df[ orders_df['customerid'] == 6878 ].head()
```

# Get a table of every single one's purchase history
> ---> 'orders' table has customerid but no names \
> ---> 'customers' has info a person and thats it \
> Merge them, so that we link _names_ w. _purchase history_

```python
orders_df.merge(customers_df, on='customerid').head()
```

```python
purchase_history_df = orders_df.merge(customers_df, on='customerid')
```

# Join : Orders-items w. orderid(/orders) and realname(/customers)
> ---> orders-items has orderid and SKU \
> ---> join _orders-items_ w. _purchase-history_ \
> and we will see which item of what id is bought by whom

```python
orders_items_df = pd.read_csv('noahs-orders_items.csv')
orders_items_df
```

```python
orders_items_df.merge(purchase_history_df, on='orderid').head()
```

```python
order_bought_by_who_df = orders_items_df.merge(purchase_history_df, on='orderid')
```

# Join : Products w. all other tables on SKU
> ---> Products and _order_bought_by_who_df_ table share a common key \
> ---> ___SKU___ \
> then we know who bought exactly what

```python
products_df = pd.read_csv('noahs-products.csv')
products_df.head()
```

```python
products_df.merge( order_bought_by_who_df, on='sku' ).head()
```

# Find all 'Noah' products

```python
finding_noah_df = products_df.merge( order_bought_by_who_df, on='sku' )

finding_noah_df[ finding_noah_df['desc'].str.contains('noah', case=False) ]
```

# Filter it, group them by personal info
> ---> group by names or phones, either one is unique anyway

```python
finding_noah_df = finding_noah_df[ finding_noah_df['desc'].str.contains('noah', case=False) ]
temp = finding_noah_df.groupby('phone')
temp.count()
```
