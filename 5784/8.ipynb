```python
import pandas as pd
import os
filenames = [_.split('.')[0] for _ in os.listdir('./') if '.csv' in _]
print(filenames)

dfs = [pd.read_csv(filename + '.csv') for filename in filenames]
df_orders, df_prod, df_order_items, df_customers = dfs
print([len(_) for _ in [df_orders, df_prod, df_order_items, df_customers]])
```

```python
customers_linked_orders = df_customers.merge(df_orders, on='customerid')
print('/len', len( customers_linked_orders ))
customers_linked_orders.head()
```

```python
order_items_linked_customers_and_orders = customers_linked_orders.merge(df_order_items, on='orderid')
print('/len', len( order_items_linked_customers_and_orders ))
order_items_linked_customers_and_orders.head()
```

```python
sku_linked_order_items = order_items_linked_customers_and_orders.merge(df_prod, on='sku')
print('/len', len(sku_linked_order_items))
sku_linked_order_items.head()
```

```python
res = sku_linked_order_items[ sku_linked_order_items['desc'].str.contains('noah', case=False) ]\
    .groupby('phone')['orderid']\
    .count()\
    .to_frame()\
    .reset_index()\
    .rename({"orderid": "merch_items"}, axis=1)\
    .sort_values("merch_items", ascending=False)
print('/len', len( res ))
res.head()
```
