```python
import pandas as pd
import os
filenames = [_.split('.')[0] for _ in os.listdir('./') if '.csv' in _]
print(filenames)

dfs = [pd.read_csv(filename + '.csv') for filename in filenames]
df_orders, df_prod, df_order_items, df_customers = dfs
print([len(_) for _ in [df_orders, df_prod, df_order_items, df_customers]])
```

```python
prices = pd.merge(df_order_items, df_prod, on='sku')
prices['margin'] = prices['qty'] * prices['unit_price'] - prices['qty'] * prices['wholesale_cost']
prices.head()
# df['profit'] = df.qty * (df.unit_price - df.wholesale_cost)
```

```python
res = pd.merge(prices, df_orders, on='orderid')
res = pd.merge(res, df_customers, on='customerid')
res.groupby(['customerid','name','phone','birthdate','address','citystatezip']).agg({'margin': 'sum'}).sort_values(by='margin').head()
```
